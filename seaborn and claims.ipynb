```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
T1=pd.read_csv('Claims2012-2014_and_weather.csv', index_col='date')
```

```python
T1['weekday']=pd.to_datetime(T1.index).weekday_name
```

```python
sns.lmplot(x='precipitation_mm', y='claims_number', hue='weekday', data=T1)
```

```python
sns.lmplot(x='t_mean_c', y='claims_number', hue='weekday', data=T1)
```

```python
sns.residplot(x='t_mean_c', y='claims_number', data=T1, color='green')
```

```python
plt.scatter(T1['t_mean_c'],T1['claims_number'], color='red', marker='o')
```

```python
sns.lmplot(x='t_mean_c', y='claims_number', row='weekday', data=T1)
```

```python
sns.violinplot(x='weekday', y='claims_number', data=T1)
```

```python
sns.boxplot(x='weekday', y='claims_number', data=T1)
```

```python
sns.jointplot(x='t_mean_c', y='claims_number', data=T1)
```

```python
sns.jointplot(x='t_mean_c', y='claims_number', data=T1, kind='kde')
```

```python
sns.pairplot(T1)
```

```python
sns.pairplot(T1, hue='weekday')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter( T1['t_mean_c'], T1['precipitation_mm'], T1['claims_number'], color='r', marker='o')
ax.set_xlabel('t_mean_c')
ax.set_ylabel('precipitation_mm')
ax.set_zlabel('claims_number')
plt.show()
```
